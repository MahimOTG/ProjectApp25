```python
# üöÄ MARS CRATER DETECTION USING RLE MASKS FROM result.json
# No external mask PNGs needed ‚Äî masks are inside result.json!

# 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 2. Install required packages
!pip install torch torchvision albumentations pycocotools opencv-python

# 3. Imports
import os
import json
import numpy as np
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision.models.segmentation import deeplabv3_resnet50
import albumentations as A
from albumentations.pytorch import ToTensorV2
from pycocotools import mask as coco_mask

# 4. Paths (Update if your folders differ)
IMG_DIR = "/content/drive/MyDrive/mars_crater_project/processed_pngs"
MASK_DIR = "/content/drive/MyDrive/mars_crater_project/annotations/mask1"

def rle_decode(rle, shape):
    """
    Safely decode RLE to binary mask.
    Handles odd-length RLE by ignoring last element.
    """
    height, width = shape
    try:
        # Ensure even length
        if len(rle) % 2 != 0:
            print(f"‚ö†Ô∏è RLE has odd length {len(rle)} ‚Äî dropping last element")
            rle = rle[:-1]  # remove last item to make even

        mask = np.zeros(height * width, dtype=np.uint8)

        for i in range(0, len(rle), 2):
            start = rle[i]
            count = rle[i + 1]
            end = start + count
            if end <= len(mask):
                mask[start:end] = 1
            else:
                mask[start:] = 1  # handle overflow

        return mask.reshape(shape)

    except Exception as e:
        print(f"‚ùå RLE decode failed: {e}")
        return np.zeros(shape, dtype=np.uint8)

# 5. RLE-Aware CraterDataset
class CraterDataset(Dataset):
    def __init__(self, img_dir, mask_dir, transform=None):
        super(CraterDataset, self).__init__()  # ‚Üê Important!
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.transform = transform
        self.pairs = []

        # Load JSON
        json_path = os.path.join(mask_dir, "result.json")
        with open(json_path, 'r') as f:
            data = json.load(f)

        for item in data:
            try:
                # Extract image name
                raw_name = item['data']['image'].split('/')[-1]
                clean_name = raw_name.split('-', 1)[1].replace('.tif', '.png')
                img_path = os.path.join(img_dir, clean_name)

                if not os.path.exists(img_path):
                    print(f"‚ö†Ô∏è Not found: {img_path}")
                    continue

                # Find crater RLE
                rle_data = None
                h, w = None, None
                for ann in item.get('annotations', []):
                    for res in ann.get('result', []):
                        if 'crater' in [lbl.lower() for lbl in res['value'].get('brushlabels', [])]:
                            rle_data = res['value']['rle']
                            h, w = res['original_height'], res['original_width']
                            break
                    if rle_data is not None:
                        break
                if rle_data is None:
                    continue

                # Decode mask
                mask = rle_decode(rle_data, (h, w)).astype(np.float32)
                self.pairs.append({'img_path': img_path, 'mask': mask})

            except Exception as e:
                print(f"‚ùå Error: {e}")
                continue

        if len(self.pairs) == 0:
            raise ValueError("No valid annotations found!")
        print(f"‚úÖ Loaded {len(self.pairs)} crater masks.")

    def __len__(self):
        return len(self.pairs)  # ‚úÖ MUST BE HERE

    def __getitem__(self, idx):
        item = self.pairs[idx]
        image = np.array(Image.open(item['img_path']).convert('L'))
        mask = item['mask'].copy()

        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']

        image = image.repeat(3, 1, 1)  # to 3-channel
        return image, mask


# 6. Augmentation Pipeline
transform = A.Compose([
    A.Resize(512, 512),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.Rotate(limit=20, p=0.5),
    A.Normalize(mean=[0.5], std=[0.5]),
    ToTensorV2()
])

# 7. Create Dataset & DataLoader
print("üîç Loading dataset from RLE...")
dataset = CraterDataset(IMG_DIR, MASK_DIR, transform=transform)
loader = DataLoader(dataset, batch_size=4, shuffle=True)

# 8. Model: DeepLabV3+ with ResNet-50 backbone
print("üß† Building model...")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = deeplabv3_resnet50(pretrained_backbone=True)
model.classifier[4] = nn.Conv2d(256, 1, kernel_size=1)  # 1 output: crater
model = model.to(device)

# 9. Loss & Optimizer
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-4)

# 10. Training Loop
print("üî• Starting training...")

model.train()
for epoch in range(15):
    epoch_loss = 0.0
    for images, masks in loader:
        images = images.to(device)
        masks = masks.unsqueeze(1).to(device).float()

        optimizer.zero_grad()
        outputs = model(images)['out']
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    avg_loss = epoch_loss / len(loader)
    print(f"Epoch {epoch+1}/15 | Loss: {avg_loss:.4f}")

# 11. Save Model
MODEL_PATH = "/content/drive/MyDrive/mars_crater_project/models/crater_model.pth"
os.makedirs(os.path.dirname(MODEL_PATH), exist_ok=True)
torch.save(model.state_dict(), MODEL_PATH)
print(f"‚úÖ Model saved to: {MODEL_PATH}")
```

```python
import numpy as np
from sklearn.metrics import f1_score, precision_score, recall_score
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
from PIL import Image
import os
import pandas as pd
import json

def compute_metrics(pred_mask, true_mask):
    y_true = true_mask.flatten()
    y_pred = pred_mask.flatten()

    if y_true.sum() == 0 and y_pred.sum() == 0:
        iou = 1.0
        dice = 1.0
    elif y_true.sum() == 0 or y_pred.sum() == 0:
        iou = 0.0
        dice = 0.0
    else:
        intersection = (y_true & y_pred).sum()
        union = (y_true | y_pred).sum()
        iou = intersection / union
        dice = 2 * intersection / (y_true.sum() + y_pred.sum())

    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    return {
        'IoU': iou,
        'F1': f1,
        'Precision': precision,
        'Recall': recall,
        'Dice': dice
    }
```

```python
from google.colab import drive
drive.mount('/content/drive')

import os
import json
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import albumentations as A
from albumentations.pytorch import ToTensorV2
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

```python
def rle_decode(rle, shape):
    """
    Safely decode RLE to binary mask.
    Handles odd-length RLEs by padding with zero if needed.
    """
    height, width = shape
    try:
        # Ensure even length
        if len(rle) % 2 == 1:
            print(f"‚ö†Ô∏è RLE has odd length {len(rle)} ‚Üí appending 0")
            rle = rle + [0]  # safe fix: add zero count at end

        mask_flat = np.zeros(height * width, dtype=np.uint8)
        index = 0
        for i in range(0, len(rle), 2):
            start = rle[i]
            count = rle[i + 1]
            end = start + count
            if start < len(mask_flat):
                mask_flat[start:end] = 1
            index += 1

        return mask_flat.reshape(shape)

    except Exception as e:
        print(f"‚ùå Failed to decode RLE: {e}")
        return np.zeros(shape, dtype=np.uint8)
```

```python
class CraterDataset(Dataset):
    def __init__(self, img_dir, mask_dir, transform=None):
        super().__init__()
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.transform = transform

        # Load JSON
        json_path = os.path.join(mask_dir, "result.json")
        with open(json_path, 'r') as f:
            data = json.load(f)

        self.pairs = []

        for item in data:
            try:
                # Extract image name
                raw_name = item['data']['image'].split('/')[-1]
                clean_name = raw_name.split('-', 1)[1].replace('.tif', '.png')
                img_path = os.path.join(img_dir, clean_name)

                if not os.path.exists(img_path):
                    continue

                # Find crater annotation
                rle_data = None
                h, w = None, None
                for ann in item.get('annotations', []):
                    for res in ann.get('result', []):
                        if 'crater' in [lbl.lower() for lbl in res['value'].get('brushlabels', [])]:
                            rle_data = res['value']['rle']
                            h, w = res['original_height'], res['original_width']
                            break
                    if rle_data is not None:
                        break
                if rle_data is None or h is None or w is None:
                    continue

                # Decode mask
                mask = rle_decode(rle_data, (h, w)).astype(np.float32)

                # Filter out tiny masks (<10 pixels) ‚Üí likely noise


                self.pairs.append({'img_path': img_path, 'mask': mask})

            except Exception as e:
                print(f"‚ùå Error: {e}")
                continue

        if len(self.pairs) == 0:
            raise ValueError("No valid annotations found after filtering.")
        print(f"‚úÖ Loaded {len(self.pairs)} annotated images.")

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        item = self.pairs[idx]
        image = np.array(Image.open(item['img_path']).convert('L'))
        mask = item['mask'].copy()

        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']

        # Convert to 3-channel
        image = image.repeat(3, 1, 1)

        return image, mask
```

```python
# Paths
IMG_DIR = "/content/drive/MyDrive/mars_crater_project/processed_pngs"
MASK_DIR = "/content/drive/MyDrive/mars_crater_project/annotations/mask1"

# Transform
transform = A.Compose([
    A.Resize(512, 512),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.Rotate(limit=20, p=0.5),
    A.GaussianBlur(blur_limit=(3, 7), p=0.3),
    A.Normalize(mean=[0.5], std=[0.5]),
    ToTensorV2()
])

# Dataset & Loader
dataset = CraterDataset(IMG_DIR, MASK_DIR, transform=transform)
loader = DataLoader(dataset, batch_size=4, shuffle=True)

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# Loss: Dice + BCE
class DiceBCELoss(nn.Module):
    def __init__(self, weight=0.5):
        super().__init__()
        self.weight = weight

    def forward(self, inputs, targets):
        bce = nn.BCEWithLogitsLoss()(inputs, targets)
        probs = torch.sigmoid(inputs)
        smooth = 1e-6
        intersection = (probs * targets).sum()
        dice = 1 - (2. * intersection + smooth) / (probs.sum() + targets.sum() + smooth)
        return bce + self.weight * dice

criterion = DiceBCELoss(weight=0.5)
optimizer = optim.AdamW(model.parameters(), lr=1e-4)
```

```python
metrics_list = []
# üîΩ Place this anywhere in your evaluation code (not inside the model)
def adaptive_threshold(prob_mask, top_k_percent=1.0):
    """
    Keep only the top K% highest confidence pixels.
    Useful for detecting small objects like craters.
    """
    flat = prob_mask.flatten()
    threshold = np.percentile(flat, 100 - top_k_percent)
    return (prob_mask > threshold).astype(bool)

for idx, item in enumerate(dataset.pairs):
    try:
        img_path = item['img_path']
        true_mask_bool = (item['mask'] > 0.5).astype(bool)

        # Load and predict
        img = Image.open(img_path).convert('L')
        img_np = np.array(img).astype(np.float32) / 255.0
        img_tensor = torch.from_numpy(img_np).unsqueeze(0).unsqueeze(0).to(device)
        img_tensor = img_tensor.repeat(1, 3, 1, 1)

        with torch.no_grad():
            output = model(img_tensor)  # U-Net++ output
            output = F.interpolate(output, size=(1024, 1024), mode='bilinear', align_corners=False)
            prob_mask = torch.sigmoid(output).cpu().numpy()[0, 0]

        # Adaptive threshold: keep top 2% most confident pixels
        threshold = np.percentile(prob_mask, 98)
        pred_mask = adaptive_threshold(prob_mask, top_k_percent=1.0)  # only top 1%

        # Compute metrics
        metrics = compute_metrics(pred_mask, true_mask_bool)
        metrics['filename'] = os.path.basename(img_path)
        metrics_list.append(metrics)

    except Exception as e:
        print(f"Error on {img_path}: {e}")
        continue

print(f"‚úÖ Evaluation complete on {len(metrics_list)} images")
```

```python
from scipy.ndimage import label
import numpy as np

def match_craters(true_mask, pred_mask, max_dist=10):
    """
    Count how many true craters have a predicted crater within `max_dist` pixels.
    """
    # Label connected components (individual craters)
    labeled_true, n_true = label(true_mask)
    labeled_pred, n_pred = label(pred_mask)

    if n_true == 0:
        return {'recall': 1.0 if n_pred == 0 else 0.0, 'precision': 1.0 if n_pred == 0 else 0.0}

    # Get centers
    if n_true > 0:
        try:
            true_centers = np.array(ndimage.center_of_mass(true_mask, labeled_true, range(1, n_true+1)))
            if n_true == 1:
                true_centers = np.array([true_centers])
        except:
            true_centers = np.empty((0, 2))
    else:
        true_centers = np.empty((0, 2))

    if n_pred > 0:
        try:
            pred_centers = np.array(ndimage.center_of_mass(pred_mask, labeled_pred, range(1, n_pred+1)))
            if n_pred == 1:
                pred_centers = np.array([pred_centers])
        except:
            pred_centers = np.empty((0, 2))
    else:
        pred_centers = np.empty((0, 2))

    # Match: for each true center, find closest pred within `max_dist`
    matched = 0
    for tc in true_centers:
        dists = np.linalg.norm(pred_centers - tc, axis=1)
        if len(dists) > 0 and dists.min() <= max_dist:
            matched += 1

    recall = matched / n_true
    precision = matched / n_pred if n_pred > 0 else 1.0 if n_true == 0 else 0.0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0

    return {'precision': precision, 'recall': recall, 'f1': f1, 'matched': matched, 'total_true': n_true}
```

```python
# 4. Load U-Net++ Model
!pip install segmentation-models-pytorch --quiet

import segmentation_models_pytorch as smp

def load_model(model_path):
    model = smp.UnetPlusPlus(
        encoder_name="resnet34",
        in_channels=3,
        classes=1
    )
    model.load_state_dict(torch.load(model_path, map_location='cpu'))
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    model.eval()
    return model, device

model, device = load_model(MODEL_PATH)
print(f"‚úÖ Model loaded on {device}")
```

```python
# 5. Adaptive Threshold Function
def adaptive_threshold(prob_mask, top_k_percent=1.0):
    flat = prob_mask.flatten()
    threshold = np.percentile(flat, 100 - top_k_percent)
    return (prob_mask > threshold).astype(bool)
```

```python
# 6. Prediction Function
def predict_craters(image_path, model, device):
    img = Image.open(image_path).convert('L')  # grayscale
    orig_size = img.size  # (width, height)
    img_np = np.array(img).astype(np.float32) / 255.0

    # To tensor: [1, 1, H, W]
    img_tensor = torch.from_numpy(img_np).unsqueeze(0).unsqueeze(0).to(device)
    img_tensor = img_tensor.repeat(1, 3, 1, 1)  # to 3-channel

    # Resize to 512x512 for model input
    img_resized = F.resize(img_tensor, size=(512, 512), interpolation=F.InterpolationMode.BILINEAR)

    # Predict
    with torch.no_grad():
        output = model(img_resized)  # [1,1,512,512]

    # Upsample back to original size
    output = F.resize(output, size=orig_size[::-1], interpolation=F.InterpolationMode.BILINEAR)
    prob_mask = torch.sigmoid(output).cpu().numpy()[0, 0]

    # Apply adaptive threshold
    pred_mask = adaptive_threshold(prob_mask, top_k_percent=1.0)

    return np.array(img), prob_mask, pred_mask
```
